```python
from sklearn.metrics import classification_report, roc_auc_score, roc_curve, precision_recall_curve, auc, f1_score,accuracy_score
from sklearn.metrics import precision_recall_curve , average_precision_score
```

```python
# function to plot the roc_curve
def plot_auc(t_y, p_y):
    fpr, tpr, thresholds = roc_curve(t_y, p_y, pos_label=1)
    fig, c_ax = plt.subplots(1,1, figsize = (9, 9))
    c_ax.plot(fpr, tpr, label = '%s (AUC:%0.2f)'  % ('classification', auc(fpr, tpr)))
    c_ax.plot([0, 1], [0, 1], color='navy', lw=1, linestyle='--')
    c_ax.legend()
    c_ax.set_xlabel('False Positive Rate')
    c_ax.set_ylabel('True Positive Rate')
```

```python
# function to plot the precision_recall_curve. You can utilizat precision_recall_curve imported above
def plot_precision_recall_curve_helper(t_y, p_y):
    fig, c_ax = plt.subplots(1,1, figsize = (9, 9))
    precision, recall, thresholds = precision_recall_curve(t_y, p_y, pos_label=1)
    aps = average_precision_score(t_y, p_y)
    c_ax.plot(recall, precision, label = '%s (AP Score:%0.2f)'  % ('classification', aps))
    c_ax.plot(recall, precision, color='red', lw=2)
    c_ax.legend()
    c_ax.set_xlabel('Recall')
    c_ax.set_ylabel('Precision')
```

```python
# function to plot the history
def plot_history(history):
    f = plt.figure()
    f.set_figwidth(15)

    f.add_subplot(1, 2, 1)
    plt.plot(history.history['val_loss'], label='val loss')
    plt.plot(history.history['loss'], label='train loss')
    plt.legend()
    plt.title("Modell Loss")

    f.add_subplot(1, 2, 2)
    plt.plot(history.history['val_accuracy'], label='val accuracy')
    plt.plot(history.history['accuracy'], label='train accuracy')
    plt.legend()
    plt.title("Modell Accuracy")

    plt.show()
```

```python
hist = plot_history(history)
```

```python

plot_auc(y_test, model_ann.predict(X_test, verbose = True))
```

```python
plot_precision_recall_curve_helper(y_test, model_ann.predict(X_test, verbose = True))
```

```python

def calc_f1(prec,recall):
    return 2*(prec*recall)/(prec+recall) if recall and prec else 0

precision, recall, thresholds = precision_recall_curve(y_test, model_ann.predict(X_test, verbose = True))
f1score = [calc_f1(precision[i],recall[i]) for i in range(len(thresholds))]
idx = np.argmax(f1score)
threshold = thresholds[idx]
print('********************************************************************************************************')
print('Precision: '+ str(precision[idx]))
print('Recall: '+ str(recall[idx]))
print('Threshold: '+ str(thresholds[idx]))
print('F1 Score: ' + str(f1score[idx]))
```

```python
accuracy_ann = accuracy_score(y_test, y_pred_ann)
print("Artificial Neural Network (ANN) Accuracy Score:", accuracy_ann)
```

# LSTM Model

```python
# Splitting into train and test sets for LSTM

X_lstm = X_scaled.reshape(X_scaled.shape[0], 1, X_scaled.shape[1])
X_train_lstm, X_test_lstm, y_train_lstm, y_test_lstm = train_test_split(X_lstm, y, test_size=0.2, random_state=42)
```

```python
# Build Model
model_lstm = Sequential()
model_lstm.add(LSTM(units=64, input_shape=(X_train_lstm.shape[1], X_train_lstm.shape[2])))
model_lstm.add(Dropout(0.5))
model_lstm.add(Dense(units=1, activation='sigmoid'))
```

```python
# Compile the Model

model_lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model

history_lstm = model_lstm.fit(X_train_lstm, y_train_lstm, epochs=30, batch_size=32, verbose=0, validation_data=(X_test_lstm, y_test_lstm))
```

```python
#Visualize the models loss and accuracy
plt.plot(history_lstm.history["accuracy"])
plt.plot(history_lstm.history["loss"])
plt.title("Model accuracy & Loss")
plt.ylabel("accuracy and loss")
plt.xlabel("epoch")
```

```python
y_pred_lstm = (model_lstm.predict(X_test_lstm) > 0.5).astype(int)

# Create a DataFrame to display the actual and predicted values
results_df = pd.DataFrame({
    'Actual': y_test_lstm.values.flatten(),
    'Predicted': y_pred_lstm.flatten()
})

# Print the first few rows of the results
print("Actual vs Predicted Values:")
print(results_df.head())
```

```python
accuracy_lstm = accuracy_score(y_test_lstm, y_pred_lstm)
print("Long Short-Term Memory (LSTM) Model Accuracy Score:", accuracy_lstm)
```

# CNN Model

```python
# Splitting into train and test sets for LSTM

X_cnn = X_scaled.reshape(X_scaled.shape[0], X_scaled.shape[1], 1)
X_train_cnn, X_test_cnn, y_train_cnn, y_test_cnn = train_test_split(X_cnn, y, test_size=0.2, random_state=42)
```

```python
# Build Model

model_cnn = Sequential()
model_cnn.add(Conv1D(32, 3, activation='relu', input_shape=(X_train_cnn.shape[1], X_train_cnn.shape[2])))
model_cnn.add(MaxPooling1D(2))
model_cnn.add(Conv1D(64, 3, activation='relu'))
model_cnn.add(MaxPooling1D(2))
model_cnn.add(Flatten())
model_cnn.add(Dense(128, activation='relu'))
model_cnn.add(Dropout(0.5))
model_cnn.add(Dense(1, activation='sigmoid'))

```

```python
# Compile the model

model_cnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model

history_cnn = model_cnn.fit(X_train_cnn, y_train_cnn, epochs=30, batch_size=32, verbose=0, validation_data=(X_test_cnn, y_test_cnn))
```

```python
#Visualize the models loss and accuracy
plt.plot(history_cnn.history["accuracy"])
plt.plot(history_cnn.history["loss"])
plt.title("Model accuracy & Loss")
plt.ylabel("accuracy and loss")
plt.xlabel("epoch")
```

```python
y_pred_cnn = (model_cnn.predict(X_test_cnn) > 0.5).astype(int)

# Create a DataFrame to display the actual and predicted values
results_df = pd.DataFrame({
    'Actual': y_test_cnn.values.flatten(),
    'Predicted': y_pred_cnn.flatten()
})

# Print the first few rows of the results
print("Actual vs Predicted Values:")
print(results_df.head(10))
```

```python

accuracy_cnn = accuracy_score(y_test_cnn, y_pred_cnn)
print("Convolutional Neural Network (CNN) Model Accuracy Score:", accuracy_cnn)
```
